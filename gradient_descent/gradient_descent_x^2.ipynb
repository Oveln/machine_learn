```python
listx = []
# The funciton is x^2+x
def gradientFun(x):
    return 2*x+1

def targetFun(x):
    return x**2 + x

def gradientDescent(starting_x, learning_rate, epsilon, max_iterations):
    x = starting_x
    for i in range(max_iterations):
        listx.append(x)
        grad = gradientFun(x)
        new_x = x - learning_rate * grad
        if grad < epsilon:
            break
        x = new_x
    return x, targetFun(x)
```

```python

print(gradientDescent(200, 0.01, 0.00001, 1000))
```
